```python
import tkinter as tk
import numpy as np
import random
Adata=[]
Bdata=[]
A=np.empty((1,1))
B=np.empty((1,1))
slect=0
window=tk.Tk()                                                         #creat the GUI
window.title("矩陣計算機")
Fsize=tk.Frame()
Fsize.grid(column=0,row=0)
Fdata=tk.Frame()
Fdata.grid(column=0,row=1)
Fbutton=tk.Frame()
Fbutton.grid(column=0,row=2)
Fcac=tk.Frame()
Fcac.grid(column=0,row=3)
Fb=tk.Frame()
Fb.grid(column=0,row=4)
Fbutton2=tk.Frame()
Fbutton2.grid(column=0,row=5)
Fans=tk.Frame()
Fans.grid(column=0,row=6)
lans=tk.Label()
```

```python
def Get_size():
    bsize['state']=tk.DISABLED
    AN=int(An.get())
    AM=int(Am.get())
    lAdata=tk.Label(Fdata,text="請輸入A矩陣內容",height=2)               #step 2nd: input the elements of the matrix A
    if AM%2==1:
        lAdata.grid(column=AM//2,row=0)
    else:
        lAdata.grid(column=AM//2-1,row=0,columnspan=2)
    for i in range(0,AN*AM):
        Adata.append(tk.Entry(Fdata,width=max(60//AM,15)))
        Adata[i].grid(column=i%AM,row=i//AM+1,sticky=tk.NW)
    bzero=tk.Button(Fbutton,text="填入0",width=10,command=ZeroA)
    bzero.grid(column=0,row=0,sticky=tk.NW)
    brand=tk.Button(Fbutton,text="填入隨機數",width=10,command=RandA)
    brand.grid(column=1,row=0,sticky=tk.NW)
    berase=tk.Button(Fbutton,text="清除全部數字",width=10,command=EraseA)
    berase.grid(column=2,row=0,sticky=tk.NW)
    bi=tk.Button(Fbutton,text="單位矩陣",width=10,command=IA)
    bi.grid(column=3,row=0,sticky=tk.NW)
    if AN!=AM:
        bi['state']=tk.DISABLED
    bgetA=tk.Button(Fbutton,text="確定",width=30,command=Get_A)
    bgetA.grid(column=0,row=1,columnspan=4)
    return
```

```python
def ZeroA():                                                            #fill all elements with 0
    AN=int(An.get())
    AM=int(Am.get())
    for i in range(0,AN*AM):
        Adata[i].delete(0,"end")
        Adata[i].insert(0,'0')
    return
def RandA():                                                            #fill all elements with random number between 1 and 100
    AN=int(An.get())
    AM=int(Am.get())
    for i in range(0,AN*AM):
        x=random.randint(1,100)
        Adata[i].delete(0,"end")
        Adata[i].insert(0,str(x))
    return
def EraseA():                                                           #delete all data in the matrix
    AN=int(An.get())
    AM=int(Am.get())
    for i in range(0,AN*AM):
        Adata[i].delete(0,"end")
def IA():                                                               #make a identity matrix
    AM=int(Am.get())                                                    #only works when N=M
    for i in range(0,AM*AM):
        Adata[i].delete(0,"end")
        if i%AM==i//AM:
            Adata[i].insert(0,'1')
        else :
            Adata[i].insert(0,'0')
    return
```

```python
def Get_A():                                                           #step 3rd: choose a option to caculate
    AN=int(An.get())
    AM=int(Am.get())
    BN=int(Bn.get())
    BM=int(Bm.get())
    global A
    A=np.resize(A,(AN,AM))
    for i in range(0,AN*AM):
        A[i//AM][i%AM]=int(Adata[i].get())
    badd=tk.Button(Fcac,text="加",width=10,command=Add)
    badd.grid(column=0,row=0)
    bmul=tk.Button(Fcac,text="乘",width=10,command=Mul)
    bmul.grid(column=1,row=0)
    bdet=tk.Button(Fcac,text="行列式值",width=10,command=Det)
    bdet.grid(column=2,row=0)
    binv=tk.Button(Fcac,text="反矩陣",width=10,command=Inv)
    binv.grid(column=3,row=0)
    if AN!=BN or AM!=BM:                                               #the four option works only in the right condition
        badd['state']=tk.DISABLED                                      #sometimes all four options are disabled
    if AM!=BN:                                                         #when it happens, press "stop" button to colse the GUI
        bmul['state']=tk.DISABLED
    if AN!=AM:
        bdet['state']=tk.DISABLED
        binv['state']=tk.DISABLED
    return
```

```python
def Fill_B(N,M):                                                        #step fourth: input the elements of the matrix B
    BN=int(Bn.get())
    BM=int(Bm.get())
    lBdata=tk.Label(Fb,text="請輸入B矩陣內容",height=2)
    if BM%2==1:
        lBdata.grid(column=BM//2,row=0)
    else:
        lBdata.grid(column=BM//2-1,row=0,columnspan=2)
    for i in range(0,N*M):
        Bdata.append(tk.Entry(Fb,width=max(60//M,15)))
        Bdata[i].grid(column=i%M,row=i//M+1,sticky=tk.NW)
    bzero=tk.Button(Fbutton2,text="填入0",width=10,command=ZeroB)
    bzero.grid(column=0,row=BN+1,sticky=tk.NW)
    brand=tk.Button(Fbutton2,text="填入隨機數",width=10,command=RandB)
    brand.grid(column=1,row=BN+1,sticky=tk.NW)
    berase=tk.Button(Fbutton2,text="清除全部數字",width=10,command=EraseB)
    berase.grid(column=2,row=BN+1,sticky=tk.NW)
    bi=tk.Button(Fbutton2,text="單位矩陣",width=10,command=IB)
    bi.grid(column=3,row=BN+1,sticky=tk.NW)
    if BN!=BM:
        bi['state']=tk.DISABLED
    bcac=tk.Button(Fbutton2,text="計算",width=30,command=Caculate)
    bcac.grid(column=0,row=BN+2,columnspan=4)
    return
```

```python
def ZeroB():                                                            #fill all elements with 0
    BN=int(Bn.get())
    BM=int(Bm.get())
    for i in range(0,BN*BM):
        Bdata[i].delete(0,"end")
        Bdata[i].insert(0,'0')
    return
def RandB():                                                            #fill all elements with random number between 1 and 100
    BN=int(Bn.get())
    BM=int(Bm.get())
    for i in range(0,BN*BM):
        x=random.randint(1,100)
        Bdata[i].delete(0,"end")
        Bdata[i].insert(0,str(x))
    return
def EraseB():                                                           #delete all data in the matrix
    BN=int(Bn.get())
    BM=int(Bm.get())
    for i in range(0,BN*BM):
        Bdata[i].delete(0,"end")
def IB():                                                               #make a identity matrix
    BM=int(Bm.get())                                                    #only works when N=M
    for i in range(0,BM*BM):
        Bdata[i].delete(0,"end")
        if i%BM==i//BM:
            Bdata[i].insert(0,'1')
        else :
            Bdata[i].insert(0,'0')
    return
```

```python
def Caculate():
    BN=int(Bn.get())
    BM=int(Bm.get())
    global A
    global B
    B=np.resize(B,(BN,BM))
    for i in range(0,BN*BM):
        B[i//BM][i%BM]=Bdata[i].get()
    global slect
    global lans
    if slect==1:
        lans=tk.Label(Fans,text=str(A+B),width=60)
        lans.grid(column=0,row=0)
        return
    if slect==2:
        lans=tk.Label(Fans,text=str(A.dot(B)),width=60)
        lans.grid(column=0,row=0)
        return
```

```python
def Add():
    BN=int(Bn.get())
    BM=int(Bm.get())
    Fill_B(BN,BM)
    global slect
    slect=1
    return
def Mul():
    BN=int(Bn.get())
    BM=int(Bm.get())
    Fill_B(BM,BN)
    global slect
    slect=2
    return
def Det():
    AM=int(Am.get())
    da=round(np.linalg.det(A),2)
    lans=tk.Label(Fcac,text=str(da),width=50,height=AM)
    lans.grid(column=0,row=1,columnspan=4)
    return
def Inv():
    AM=int(Am.get())
    global lans
    if np.linalg.det(A):
        inv=np.linalg.inv(A)
        lans=tk.Label(Fcac,text=str(inv),width=60)
        lans.grid(column=0,row=1,columnspan=4)
    else:
        lans=tk.Label(Fcac,text="反矩陣不存在",width=60,height=AM)
        lans.grid(column=0,row=1,columnspan=4)
    return
```

```python
def Stop():                                                             #close the GUI
    window.destroy()
    return
```

```python
lAsize=tk.Label(Fsize,text="請輸入A矩陣大小",height=2)                   #step 1st: input the size of the matrix
lAsize.grid(column=0,row=0,columnspan=2)
lBsize=tk.Label(Fsize,text="請輸入B矩陣大小",height=2)
lBsize.grid(column=2,row=0,columnspan=2)
lAn=tk.Label(Fsize,text="行數")
lAn.grid(column=0,row=1)
An=tk.Entry(Fsize,width=10)
An.grid(column=0,row=2)
lAm=tk.Label(Fsize,text="列數")
lAm.grid(column=1,row=1)
Am=tk.Entry(Fsize,width=10)
Am.grid(column=1,row=2)
lBn=tk.Label(Fsize,text="行數")
lBn.grid(column=2,row=1)
Bn=tk.Entry(Fsize,width=10)
Bn.grid(column=2,row=2)
lBm=tk.Label(Fsize,text="列數")
lBm.grid(column=3,row=1)
Bm=tk.Entry(Fsize,width=10)
Bm.grid(column=3,row=2)
bsize=tk.Button(Fsize,text="確定",height=3,width=25,command=Get_size)
bsize.grid(column=0,row=3,columnspan=4)
bend=tk.Button(window,text="結束",height=2,width=15,command=Stop)
bend.grid(column=0,row=100)
window.mainloop()
```
